1.1.2 전일 대비 감염률 증감 기준 시기 구분

1.1.1의 가시화된 그래프를 살펴보면, 발생 추이에 따른 시기는 확산세가 유지되며 일정 수준의 감염률을 보이는 급간과 높은 확산세를 보이며 감염률이 큰 폭으로 증가하는 두 급간이 눈에 띄게 구분된다. 즉, 가시적으로 나타나는 감염률 차이에 기반하여 비교ㆍ대조에 이용할 시기를 나눌 수 있을 것으로 보인다.

 그러나 좀 더 설득력있고 명확한 구분을 위해 <u>*사이킷런 패키지*</u> 의 선형회귀 모델을 이용하여 급간을 구분하기로 했다.

```python
# 확산세 기울기(전일 대비 감염률 증가)를 반영한 코로나 19 확진자 데이터

plt.figure(figsize=(12,6))
plt.plot(range(133),time_p['confirmed'],'bo',alpha=0.5,linewidth=3)
plt.ylim([0,1100])
plt.xticks(7*np.arange(19),weeks,rotation=30)
plt.title('서울시_코로나19_발생추이',size=20)
plt.axvline(7*5, c='r',linestyle='--',alpha=0.8)
plt.axvline(7*11, c='r',linestyle='--',alpha=0.8) 
plt.axvline(7*16, c='r',linestyle='--',alpha=0.8) 

# 회귀선을 그리기 위한 LinearRegression import
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
x_train = np.arange(34,77).reshape(-1,1)
x_test = time_p.loc[34:76]['confirmed'].to_numpy().reshape(-1,1)
lr.fit(x_train,x_test)
print('Trend_Line2: {0:.4f}x+{1:.4f}'.format(lr.coef_[0][0],lr.intercept_[0]))
print('Trend_Line2 구간의 전일 대비 증가량 평균 : {}명'.format(round(lr.coef_[0][0])))
plt.plot([34,77],[34*lr.coef_[0][0]+lr.intercept_[0],77*lr.coef_[0][0]+lr.intercept_[0]],linewidth=5,color='m',alpha=0.8)

lr = LinearRegression()
x_train = np.arange(111,133).reshape(-1,1)
x_test = time_p.loc[111:132]['confirmed'].to_numpy().reshape(-1,1)
lr.fit(x_train,x_test)
print('\nTrend_Line4: {0:.4f}x+{1:.4f}'.format(lr.coef_[0][0],lr.intercept_[0]))
print('Trend_Line4 구간의 전일 대비 증가량 평균 : {}명'.format(round(lr.coef_[0][0])))
plt.plot([111,132],[111*lr.coef_[0][0]+lr.intercept_[0],132*lr.coef_[0][0]+lr.intercept_[0]],linewidth=5,color='m',alpha=0.8)

plt.text(7*1+3,950,'Trend1',size=15)
plt.text(7*7+3,950,'Trend2',size=15)
plt.text(7*13,950,'Trend3',size=15)
plt.text(7*17+2,950,'Trend4',size=15)

plt.text(45,200,'전일 대비 증가량 평균 : 12명',rotation=38,size=10,color='m')
plt.text(113,600,'전일 대비 증가량 평균 : 8명',rotation=28,size=10,color='m')

plt.show()
```

 그 결과 코로나 19 발생 추이의 선형 방정식에 따라 급간을 Trend1, Trend2, Trend3, Trend4 이상 4개로 분할할 수 있었다. (기준 일자 2020-02-14 / 2020-04-06 / 2020-05-11)
 
 우선 Trend1과 Trend3는 전날 대비 코로나 19 일일 감염률 증가가 0에 수렴하는 구간으로, ‘전염 소강 구간’으로 판단하였다. 그리고 Trend2와 Trend4는 감염률이 선형적으로 증가하는 구간으로, Trend2 구간에서는 당일 발생 추가 확진자 수가 전일 대비 평균 12명씩 증가하였으며, Trend4 구간에서는 8명씩 증가하는 것을 알 수 있었다.

전체 분석 대상 기간을 구분하는 세 가지 기준의 유효성을 평가하기 위해 서울시내 배달 관련 데이터와 비교하여 연관 정도를 파악하고자 한다. 이를 위해 세 가지 기준을 설정했다.

- 선형 흐름의 일치 정도 비교
- 배달량 평균 비교 (급격한 차이 = 좋은 구분)
- 급간 변화 구간 전후 비교

1.2.1 그래프 개형 일치 파악

먼저 서울시내 배달량 변화 흐름과 각 급간의 확진자 증감 흐름이 가장 잘 일치하는 구분을 선정하여 가장 유의미한 인사이트 도출을 이끌어낼 수 있는 기준을 선정할 것이다. 이를 위해 서울시내 일일 배달 주문량을 가시화했다.


```python
# 배달 호출 데이터 load
deliv_data = pd.read_csv('data/dlvr_call.csv')

# 파생변수 생성 - year, month 칼럼 생성
deliv_data['PROCESS_DT']=pd.to_datetime(deliv_data['PROCESS_DT'])
deliv_data['year'] = pd.to_datetime(deliv_data['PROCESS_DT'].values).year
deliv_data['month'] = pd.to_datetime(deliv_data['PROCESS_DT'].values).month

# 결측치 처리
deliv_data['SETLE_CARD_NM'].fillna('',inplace=True)

# 미사용 칼럼 처리.
deliv_data.drop(['DLVR_REQUST_ENTRPS_ID','DLVR_RCEPT_ENTRPS_ID','DLVR_STORE_ID'],axis=1,inplace=True)
deliv_data.drop(['DLVR_STORE_LEGALDONG_CODE','DLVR_STORE_LI_NM','DLVR_STORE_ADSTRD_CODE','DLVR_STORE_RDNMADR_CODE'],axis=1,inplace=True)
deliv_data.drop(['DLVR_DSTN_LEGALDONG_CODE','DLVR_DSTN_LI_NM','DLVR_DSTN_ADSTRD_CODE','DLVR_DSTN_RDNMADR_CODE'],axis=1,inplace=True)
deliv_data.drop(['DLVR_MAN_ID'],axis=1,inplace=True)

# 배달이 정상적으로 완료된 데이터만 선별.
deliv_data = deliv_data[deliv_data['DLVR_REQUST_STTUS_VALUE']==1]

deliv_data.head(4)
```

```python
# dlvr_call.csv 데이터 중 서울시 데이터만 추출.
seoul_deliv = deliv_data[(deliv_data['DLVR_STORE_BRTC_NM']=='서울특별시') & (deliv_data['PROCESS_DT']>=init)]
seoul_deliv = seoul_deliv.sort_values(by='PROCESS_DT')
seoul_deliv = seoul_deliv.reset_index().drop('index',axis=1)
seoul_deliv_count = seoul_deliv.groupby(['PROCESS_DT'])['DLVR_REQUST_STTUS_VALUE'].count()

plt.figure(figsize=(12,6))
plt.plot(range(132),seoul_deliv_count.values)
plt.xticks(7*np.arange(19),weeks,rotation=30)
plt.title('서울시_배달량_변화',size=20)

plt.show()
```

 가시화한 그래프를 살펴보면, 일정 주기가 반복되는 것을 확인할 수 있다. 그러나 1.1.1과의 개형 일치 정도를 비교하기에는 무리가 있다고 판단했다.

```python
seoul_deliv_count_df = pd.DataFrame(seoul_deliv_count).sort_values(by='DLVR_REQUST_STTUS_VALUE',ascending=False).rename(columns={'DLVR_REQUST_STTUS_VALUE':'배달량'})
seoul_deliv_count_df.head(4)
```

1.2.2 구분된 시기의 배달 주문량 평균 비교

가, 나, 다 기준에 따라 구분한 각 시기의 서울시내 일일 배달량 평균을 구했다. dlvr_call.csv 데이터를 통해 구한 각 시기별 배달량 평균을 아래 그래프로 가시화했다. 또한 전체 시기의 서울시 일일 배달량 평균을 구했고, 비교를 위해 마찬가지로 그래프에 표시했다.


```python
# 코로나 19 발생추이 0224 / 0406 / 0511

print(datetime.datetime(2020,2,24) - datetime.datetime(2020,1,20))
print(datetime.datetime(2020,4,6) - datetime.datetime(2020,2,24))
print(datetime.datetime(2020,5,11) - datetime.datetime(2020,4,6))
print(datetime.datetime(2020,5,31) - datetime.datetime(2020,5,11))

def divide_by_ac(x):
    r=''
    if x <= datetime.datetime(2020,2,24):
        r = 'phase1'
    elif x <= datetime.datetime(2020,4,6):
        r = 'phase2'
    elif x <= datetime.datetime(2020,5,11):
        r = 'phase3'
    else:
        r = 'phase4'
    return(r)

seoul_deliv['phase'] = seoul_deliv['PROCESS_DT'].apply(lambda x : divide_by_ac(x))
c19_by_ac = seoul_deliv.groupby(['phase'])['DLVR_REQUST_STTUS_VALUE'].count()

seoul_deliv['PROCESS_DT'].unique().shape[0]
seoul_deliv['PROCESS_DT'].shape[0]
all_mean = int(seoul_deliv['PROCESS_DT'].shape[0]/seoul_deliv['PROCESS_DT'].unique().shape[0])

plt.figure(figsize=(8,6))
plt.bar(c19_by_ac.index,c19_by_ac.values/[35,42,35,20],width=0.5, color='r', alpha=0.5)
plt.xticks(['phase1','phase2','phase3','phase4'],size=15)
plt.axhline(all_mean,color='b',linestyle='--',alpha=0.8)
plt.title('코로나19 발생추이별 전후배달량 (일 평균)',size=15)

plt.show()
```

 이 경우 역시 앞서 말한 심리적 긴장감 해소로 인한 외출 증가가 반영되었다고 판단된다. 마찬가지로 평균 산출 기간을 기준일자 직전 및 직후로 축소하여 다시 평가할 수 있을 것이다.

 끝으로 다 기준의 그래프를 살펴보자면, 방역 지침이 없는 기간인 trend1, 사회적 거리두기가 시행된 기간인 trend2, 생활속 거리두기로 격하된 trend3을 거치면서 증가 후 감소 경향을 띨 것이 예상되었다. 하지만 예상 추세가 반영되지 않고 되려  감소 후 증가한 것이 확인됐다.

 이 다 그래프의 경우 가, 나와 달리 예상 경향과 완전히 상반되는 결과가 도출되었기 때문에 가장 설득력이 떨어지는 구분이다.


1.2.3 기준 일자 전후 일주일 배달 주문량 비교

1.2.2에서 진행한 데이터 분석은 단순히 구분한 기간별 전체 평균을 비교했다. 기준이 되는 시점으로부터의 배달량 변화를 더 명확하게 파악하고자 기준일 전후 일주일의 배달 주문량 데이터를 가시화했다.

먼저 <가> 기준에 따른 전후 일주일 배달량 추이를 가시화한 그래프이다.


```python
# 코로나 19 발생추이 0224 / 0406 / 0511 
# 감염률 증가ㆍ유지 시기 구분

ac_p1 = seoul_deliv[ (datetime.datetime(2020,2,17) <= seoul_deliv['PROCESS_DT']) &  ( seoul_deliv['PROCESS_DT'] <= datetime.datetime(2020,3,2))].groupby(['PROCESS_DT'])['DLVR_REQUST_STTUS_VALUE'].count()
ac_p2 = seoul_deliv[ (datetime.datetime(2020,3,30) <= seoul_deliv['PROCESS_DT']) &  ( seoul_deliv['PROCESS_DT'] <= datetime.datetime(2020,4,13))].groupby(['PROCESS_DT'])['DLVR_REQUST_STTUS_VALUE'].count()
ac_p3 = seoul_deliv[ (datetime.datetime(2020,5,4) <= seoul_deliv['PROCESS_DT']) &  ( seoul_deliv['PROCESS_DT'] <= datetime.datetime(2020,5,18))].groupby(['PROCESS_DT'])['DLVR_REQUST_STTUS_VALUE'].count()

plt.figure(figsize=(20,4))

plt.subplot(1,3,1)
plt.plot(ac_p1.index,ac_p1.values)
plt.axvline(datetime.datetime(2020,2,24), c='r',linestyle='--',alpha=0.8)
plt.xticks(rotation=30)
plt.title('Trend1 일주일 전후 배달량')

plt.subplot(1,3,2)
plt.plot(ac_p2.index,ac_p2.values)
plt.axvline(datetime.datetime(2020,4,6), c='r',linestyle='--',alpha=0.8)
plt.xticks(rotation=30)
plt.title('Trend2 일주일 전후 배달량')

plt.subplot(1,3,3)
plt.plot(ac_p3.index,ac_p3.values)
plt.axvline(datetime.datetime(2020,5,11), c='r',linestyle='--',alpha=0.8)
plt.xticks(rotation=30)
plt.title('Trend3 일주일 전후 배달량')

plt.show()
```

 <가> 기준 설정 방식은 가시화된 감염률 그래프인 1.1.1 형태를 감염세 증가와 유지로 나눠 구분한 것이다. 기준 전후 일주일 동안의 서울시내 배달량을 살폈을 때, 배달량 증감이 감염률 추이를 따라 이뤄진다면 <가> 기준이 유의미하다는 것을 뜻한다. ‘배달량이 감염률에 비례한다'는 가설을 검증하기 위해 기준을 설정했기 때문이다.

 먼저 Trend 1과 2를 구분한 첫 그래프의 경우, 같은 주기 형태이지만 눈에 띄게 배달량이 증가했다. 확진자가 늘어나기 시작한 시점이라는 것을 상기한다면, 배달량이 함께 늘어난 이 결과는 의미가 크다. ‘감염률이 늘어나면 배달 음식을 이용하는 사람이 늘어날 것이다'라는 가설에 부합하기 때문이다.
 
 Trend 2와 3 시기를 구분한 두 번째 그래프의 전후 일주일은 같은 주기 형태로 비슷한 배달량 수준을 유지했다. 이 역시 유의미한 결과인데, Trend 2와 3을 구분하는 기준은 감염률이 꾸준히 증가하다가 확산세를 유지하며 일정한 감염률에 접어드는 기점이기 때문이다. 즉, 기준 이전 증가하던 서울시내 감염률이 일정 수준을 유지하기 시작했다. 감염률이 더 이상 증가하지 않게 된 시기에 배달량이 일정하게 유지되었으므로, 마찬가지로 ‘감염률과 배달량이 비례한다'는 가설에 부합한다.

 마지막으로 Trend 3과 4를 구분한 그래프는 뚜렷한 특징을 찾기 힘들다. 주기 형태도 다소 무너졌으며 가설에 따라 배달량이 증가했어야 하는 시점임에도 오히려 감소했음을 알 수 있다. 이런 결과가 나온 데에는 다양한 요인이 있을 것이다. 이미 서울시내 배달량이 일정 수준 이상, 즉 포화 상태에 다다라 기간을 축소해서 진행한 분석의 의미가 옅어졌을 수 있다.

 결과적으로 <가> 구분 기준의 첫 번째와 두 번째 기준에서는 전후 일주일 동안 ‘서울시내 감염률과 배달량은 비례한다’는 가설을 뒷받침하는 결과를 얻었다. 반면 세 번째 그래프에서는 가설과 다른 결과가 나왔다.

2.1.1 서울시 행정동별 배달 주문량 증감 분석

먼저 'dlvr_call.csv'를 행정동 단위로 구분하여 각 phase 별 배달 음식 주문량을 구해 표로 나타냈다.

```python
# dlvr_call.csv 데이터 급간 나누기.

phase1 = seoul_deliv[ seoul_deliv['PROCESS_DT'] <= datetime.datetime(2020,2,17)].groupby(['DLVR_STORE_LEGALDONG_NM'])['DLVR_REQUST_STTUS_VALUE'].count()
phase2 = seoul_deliv[ (datetime.datetime(2020,2,18) <= seoul_deliv['PROCESS_DT']) & (seoul_deliv['PROCESS_DT'] <= datetime.datetime(2020,3,9))].groupby(['DLVR_STORE_LEGALDONG_NM'])['DLVR_REQUST_STTUS_VALUE'].count()
phase3 = seoul_deliv[ (datetime.datetime(2020,3,10) <= seoul_deliv['PROCESS_DT']) & (seoul_deliv['PROCESS_DT'] <= datetime.datetime(2020,5,11))].groupby(['DLVR_STORE_LEGALDONG_NM'])['DLVR_REQUST_STTUS_VALUE'].count()
phase4 = seoul_deliv[ datetime.datetime(2020,5,12) <= seoul_deliv['PROCESS_DT']].groupby(['DLVR_STORE_LEGALDONG_NM'])['DLVR_REQUST_STTUS_VALUE'].count()

print(datetime.datetime(2020,2,17) - datetime.datetime(2020,1,20))
print(datetime.datetime(2020,3,9) - datetime.datetime(2020,2,17))
print(datetime.datetime(2020,5,11) - datetime.datetime(2020,3,9))
print(datetime.datetime(2020,5,31) - datetime.datetime(2020,5,11),'\n')

phase1 = pd.DataFrame(phase1).reset_index().rename(columns={'DLVR_REQUST_STTUS_VALUE':'phase1'})
phase2 = pd.DataFrame(phase2).reset_index().rename(columns={'DLVR_REQUST_STTUS_VALUE':'phase2'})
phase3 = pd.DataFrame(phase3).reset_index().rename(columns={'DLVR_REQUST_STTUS_VALUE':'phase3'})
phase4 = pd.DataFrame(phase4).reset_index().rename(columns={'DLVR_REQUST_STTUS_VALUE':'phase4'})

phases = [phase1,phase2,phase3,phase4]

total = pd.DataFrame()
date = [28,21,63,20]
for i,phase in enumerate(phases):
    phase[f'phase{i+1}'] = phase[f'phase{i+1}']/date[i]
    phase = phase.sort_values(by=f'phase{i+1}',ascending=False)
    phase[f'phase{i+1}'] = phase[f'phase{i+1}'].astype(int)
    phase = phase.head(8)
    if i==0 :
        total = phase
    else :
        total = pd.merge(total,phase,how='outer')

# 결측치 처리.
print(total.isna().sum())
total.fillna(0,inplace=True)

total
```

서울시에는 400개가 넘는 행정동이 있지만, (데이터)는 총 65개 행정동의 배달 데이터를 포함한다. 따라서 구 단위로 분석 범위를 늘리면 분석 결과를 신뢰하기 힘들다. 또한, 배달 주문량 하위 그룹은 배달 주문량이 10건도 채 되지 않는다.

 따라서 분석의 정확성과 의의를 확보하기 위하여 전체 배달량이 많은 순서대로 행정동을 추려 분석 범위를 좁혀 진행한다. 행정동별 분석과는 별개로 서울시 전체 배달 데이터를 분석해 거시적인 인사이트를 도출함과 동시에 비교ㆍ분석의 기준으로 삼는다.

```python
# 서울 배달 데이터(seoul_deliv)에 업종별 category 파생변수 생성.

카페_디저트 = ['카페/디저트']
한식 = ['족발/보쌈','분식','한식','도시락','찜탕']
일식 = ['돈까스/일식','회']
양식 = ['치킨','아시안/양식','패스트푸드']
중식 = ['중식']
기타 = ['야식','배달전문업체','심부름']

def diving_category(x):
    if x in 카페_디저트:
        return('카페/디저트')
    elif x in 한식:
        return('한식')
    elif x in 일식:
        return('일식')
    elif x in 양식:
        return('양식')
    elif x in 중식:
        return('중식')
    elif x in 기타:
        return('기타')

seoul_deliv['category']=seoul_deliv['DLVR_STORE_INDUTY_NM'].apply(lambda x : diving_category(x))
seoul_deliv.head(4)
```

배달량은 말 그대로 정량적인 정보이다. 배달 주문량이 많다는 것에 많은 함의가 있지만, 배달 시스템을 평가하기 위해 단순히 배달량 정보만을 분석하는 것은 한계가 있다.

 따라서 우리 팀은 배달의 질을 평가할 수 있는 요소를 데이터에서 발견해 분석하고자 했다. 'dlvr_call.csv'에서 주문이 접수되고 배달이 완료될 때까지의 시간을 분석하여 서울시 전체 평균 일일 배달 소요 시간을 도출하여 그래프화하였다.

2.4.1 분석 기간 전체 배달 소요시간 추이

```python
# 결측치 처리.
seoul_deliv = seoul_deliv[seoul_deliv['DLVR_RECPT_TIME'] != '\\N']
seoul_deliv = seoul_deliv[seoul_deliv['DLVR_COMPT_TIME'] != '\\N']

# datetime 타입으로 변환 및 배달소요시간 계산.
seoul_deliv['DLVR_RCEPT_TIME']=pd.to_datetime(seoul_deliv['DLVR_RCEPT_TIME'])
seoul_deliv['DLVR_CARALC_TIME']=pd.to_datetime(seoul_deliv['DLVR_CARALC_TIME'])
seoul_deliv['DLVR_RECPT_TIME']=pd.to_datetime(seoul_deliv['DLVR_RECPT_TIME'])
seoul_deliv['DLVR_COMPT_TIME']=pd.to_datetime(seoul_deliv['DLVR_COMPT_TIME'])
seoul_deliv['배달소요시간'] = seoul_deliv['DLVR_COMPT_TIME'] -  seoul_deliv['DLVR_RCEPT_TIME']
seoul_deliv['timedelta'] = pd.to_numeric(seoul_deliv['배달소요시간'])
seoul_deliv['timedelta'] = seoul_deliv['timedelta'].values/(10**9*60)
```

## 3. 결론 Conclusion

앞서 분석 기준을 선정하고, 인사이트를 창출하는 과정에서 발견한 유의미한 결과들을 요약 정리해보겠다.

우선 첫째, 코로나 확진자 발생추이 (전일 대비 감염률 증감), 대규모 확산사건, 서울시 방역지침 시행 중 배달량에 가장 큰 영향을 끼치는 요소는 대규모 확산 사건의 발생이다.

 이 경우 대규모 확산 사건의 발생이 곧 즉각적인 보도 및 뉴스 확산으로 이어져 시민들에게 심리적으로 가장 큰 영향을 끼쳤기 때문이라고 판단된다. 또 서울시의 방역 지침 시행이 배달량과 특히 연관이 적었던 이유 중 하나는 데이터가 코로나 발생 초기 시기에 기반하였기 때문이다. 따라서 ‘O인 이상 집합 금지 조항’이나 ‘O시 이후 점포 이용 금지’와 같은 조항이 시기 내에 있었다면 배달량 증감에 큰 영향을 끼쳤을 것이다. 

둘째, 대규모 확산 사건이 발생한다고 하더라도, 그 영향이 배달량에 있어 긴 시간에 걸쳐 축적되어 나타나지는 않는다.

 즉, 전체적인 수준에서의 배달량의 변화는 크게 드러나지 않았으며 (배달량은 주로 일주일 단위로 큰 주기를 가지며 반복되었다.) , 코로나가 영향을 끼쳐 변화를 일으키는 시점은 사건 발생 전 후 일주일, 즉 총 14일간의 흐름에 주목해야 할 필요가 있었다.

셋째, 대규모 확산 사건이 발생하는 족족 비례하여 배달량이 늘어나지는 않는다. 

 사건 발생에 따른 배달량 증가에도 한계가 있는 것으로 파악되었다. 주어진 데이터를 기반으로 실시한 분석에 의거하면 신천지 사건과 신도림 콜센터 사건의 경우 비슷한 배달량 증가 및 배달 취소율, 배달 소요 시간 등 다양한 데이터 분석에서 비슷한 흐름을 보였으나, 이태원 클럽발 사건의 경우 전혀 다른 추이를 보였다.

 즉, 사건이 발생하는 족족 똑같은 흐름으로 배달업계에 영향을 끼치는 것은 아니며, 사건의 성격 및 배달 포화도에 따라 어느정도 한계가 존재한다는 사실을 파악했다.


하지만 어떤 기점을 기준으로 이 코로나 확진자 수 증가에 따른 배달업계 포화가 이루어지는지는 데이터 상으로 명확히 밝혀내지 못했다. 이 부분에 대해서는 더 자세한 분석과 고찰이 필요할 것으로 보인다. 

넷째, 대규모 사건 발생에 따라 배달량은 증가하는 흐름을 보였다.
  
 지역구 별 총 배달량이 증가했을 뿐만 아니라, 업종별로 나누어 보았을 때 역시 총 배달량은 증가하는 흐름이었다. 특히 평소 배달량이 많은 지역과 업종에 대해 그 영향력이 크게 나타났다. 주어진 데이터 상으로는 서울시 내 지역구중에서는 개봉동이, 업종 중에서는 양식이 주목할만한 변화량을 보였다. 

 하지만 개/폐업이 활발하게 이루어지는 지역구의 경우에는 되려 사건 발생이 큰 영향을 끼치지 못했고, 일정한 흐름을 보이지도 않았다. 즉, 주로 강남구, 서초구와 같은 매매 시세가 비싸고 유동 인구가 많은 지역에서는 배달량이 규칙성을 띠지 못하고 튀는 경향을 보인 것이었다.

다섯째, 배달 접수량 대비 취소량, 즉 배달 취소율은 대규모 사건 발생 직후 크게 감소하였다.

 그 이유는 대규모 확산 사건이 발생함에 따라 배달은 선택이 아닌 어느정도 불가피하게 필요한 존재로 자리잡게 되면서, 소비자들이 배달 시간이나 조건에 크게 동요하지 않게 되었음으로 추측된다.

여섯째, 배달 소요 시간은 배달량 증가와 같은 흐름을 보이되, 그 정도가 훨씬 더 예민하고 극단적으로 나타났다.
 
 배달량 자체는 대규모 확산 사건 발생에 따라 직관적으로 대응하여 증가할 수 있지만, 배달 소요 시간은 배달량 증가 시 교통체증 및 배달 라이더 포화와 같은 형태로 그 영향이 누적되어 더 극단적으로 증가할 수 밖에 없었던 것으로 추측된다.
